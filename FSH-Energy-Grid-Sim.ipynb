```python
import numpy as np
import matplotlib.pyplot as plt

#time
hours=np.arange(0,24,0.1)
```

```python
# Demand curve 
demand = 50 + 20*np.sin((hours - 7) * np.pi / 12)
```

```python
# Wind power (random + fluctuating)
np.random.seed(0)
wind = 30 + 10*np.sin(hours * np.pi / 6) + 5*np.random.randn(len(hours))
```

```python
# Graph
plt.plot(hours, demand, label="Demand")
plt.plot(hours, wind, label="Wind Supply")
plt.xlabel("Hour")
plt.ylabel("Power")
plt.legend()
plt.title("Energy Grid Simulation (Basic)")
plt.savefig("Energy-Grid-Simulation-(Basic).png")
plt.show()
```
